```python
import time
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression

import pickle
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

```python
def encode_cate_cols(dataset):

    cate_cols = dataset.select_dtypes(include='object').columns

    le = LabelEncoder()

    for col in cate_cols:
        dataset[col] = le.fit_transform(dataset[col].astype(str))

    return dataset

# dataset remains original data , encoded_df will be where categorical columns encoded
```

```python
def split_scalar(indep_X,dep_Y):
    
    X_train,X_test,y_train,y_test = train_test_split(indep_X, dep_Y , test_size = 0.25, random_state = 0)
    sc = StandardScaler()
    X_train = sc.fit_transform(X_train)
    X_test = sc.transform (X_test)
    return X_train , X_test , y_train , y_test
    
```

```python
def apply_pca(X_train , X_test,n):

    pca = PCA(n_components = n )
    X_train_pca = pca.fit_transform(X_train)
    X_test_pca  = pca.transform(X_test)
    explained_variance = pca.explained_variance_ratio_

    return X_train_pca , X_test_pca , pca
```

```python
def cm_prediction(classifier,X_test, y_test):
    
    y_pred = classifier.predict(X_test)
    
    cm = confusion_matrix(y_test , y_pred)

    Accuracy = accuracy_score(y_test , y_pred)

    report = classification_report(y_test, y_pred)
    
    return classifier,Accuracy,report,X_test,y_test,cm
    
```

```python
def logistic(X_train_pca,y_train,X_test_pca , y_test ):
    
    classifier = LogisticRegression(random_state = 0)
    classifier.fit(X_train_pca , y_train)
    classifier,Accuracy,report,X_test,y_test,cm = cm_prediction(classifier,X_test_pca,y_test)
    
    return classifier,Accuracy,report,cm , X_test , y_test
```

```python
def svm_linear(X_train_pca ,y_train, X_test_pca , y_test ):

    from sklearn.svm import SVC
    classifier = SVC(kernel = 'linear' , random_state = 0)
    classifier.fit(X_train_pca , y_train)
    classifier,Accuracy, report,X_test,y_test,cm = cm_prediction(classifier,X_test_pca,y_test)
    
    return classifier , Accuracy , report ,cm , X_test , y_test
```

```python
def svm_NL(X_train_pca, y_train ,X_test_pca , y_test  ):

    from sklearn.svm import SVC
    classifier = SVC(kernel = 'rbf',random_state = 0 )
    classifier.fit(X_train_pca,y_train)
    classifier,Accuracy,report,X_test,y_test,cm = cm_prediction(classifier,X_test_pca,y_test)
    
    return classifier,Accuracy,report,X_test,y_test,cm
   
```

```python
def Navie(X_train_pca,y_train,X_test_pca , y_test ):

    from sklearn.naive_bayes import GaussianNB
    classifier = GaussianNB()
    classifier.fit(X_train_pca , y_train)
    classifier,Accuracy,report,X_test,y_test,cm = cm_prediction(classifier,X_test_pca,y_test)
    
    return classifier,Accuracy,report,X_test,y_test,cm
```

```python
def knn(X_train_pca,y_train,X_test_pca , y_test ):
       
    from sklearn.neighbors import KNeighborsClassifier
    classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
    classifier.fit(X_train_pca, y_train)
    classifier,Accuracy,report,X_test,y_test,cm = cm_prediction(classifier,X_test_pca,y_test)
    return  classifier,Accuracy,report,X_test,y_test,cm
```

```python
def Decision(X_train_pca,y_train,X_test_pca, y_test ):

    from sklearn.tree import DecisionTreeClassifier
    classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
    classifier.fit(X_train_pca, y_train)
    classifier,Accuracy,report,X_test,y_test,cm = cm_prediction(classifier,X_test_pca,y_test)
    return  classifier,Accuracy,report,X_test,y_test,cm      
```

```python
def random(X_train_pca,y_train,X_test_pca , y_test ):
    
    from sklearn.ensemble import RandomForestClassifier
    classifier = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
    classifier.fit(X_train_pca, y_train)
    classifier,Accuracy,report,X_test,y_test,cm = cm_prediction(classifier,X_test_pca,y_test)
    return  classifier,Accuracy,report,X_test,y_test,cm
```

```python
def pca_classification(acclog,accsvml,accsvmnl,accknn,accnav,accdes,accrf): 
    
    results_df =pd.DataFrame(index=[2,3,4,5,6],columns=['Logistic','SVMl','SVMnl','KNN','Navie','Decision','Random'])

    for number, idex in enumerate(results_df.index):
        
        results_df.loc[idex, 'Logistic'] = acclog[number]         
        results_df.loc[idex,'SVMl']      = accsvml[number]
        results_df.loc[idex,'SVMnl']     = accsvmnl[number]
        results_df.loc[idex, 'KNN']      = accknn[number]            
        results_df.loc[idex, 'Navie']    = accnav[number]   
        results_df.loc[idex, 'Decision'] = accdes[number]   
        results_df.loc[idex, 'Random']   = accrf[number]   
     
     
    return results_df
```

```python
dataset = pd.read_csv('CKD.csv')

encoded_df = encode_cate_cols(dataset.copy())
```

```python
encoded_df = pd.get_dummies(dataset, drop_first=True)

indep_X = encoded_df.drop(columns=['classification_yes'])
dep_Y = encoded_df['classification_yes']
```

```python


acclog=[]
accsvml=[]
accsvmnl=[]
accknn=[]
accnav=[]
accdes=[]
accrf=[]

X_train, X_test, y_train, y_test = split_scalar(indep_X,dep_Y)   

for n in range(2,7):

    X_train_pca, X_test_pca, pca = apply_pca(X_train,X_test,n)
      
    classifier,Accuracy,report,cm ,_ ,_ = logistic(X_train_pca,y_train,X_test_pca , y_test)
    acclog.append(Accuracy)

    classifier,Accuracy,report,cm ,_ ,_ = svm_linear(X_train_pca,y_train,X_test_pca , y_test)  
    accsvml.append(Accuracy)
    
    classifier,Accuracy,report,cm ,_ ,_ = svm_NL(X_train_pca,y_train,X_test_pca , y_test)  
    accsvmnl.append(Accuracy)
    
    classifier,Accuracy,report,cm,_,_  = knn(X_train_pca,y_train,X_test_pca , y_test)  
    accknn.append(Accuracy)
    
    classifier,Accuracy,report,cm ,_ ,_ = Navie(X_train_pca,y_train,X_test_pca, y_test)  
    accnav.append(Accuracy)
    
    classifier,Accuracy,report,cm ,_ ,_ = Decision(X_train_pca,y_train,X_test_pca, y_test)  
    accdes.append(Accuracy)
    
    classifier,Accuracy,report,cm,_ ,_ = random(X_train_pca,y_train,X_test_pca, y_test)  
    accrf.append(Accuracy)


results_df = pca_classification( acclog,accsvml,accsvmnl,accknn,accnav,accdes,accrf )
```
